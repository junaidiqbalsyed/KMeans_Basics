# DBSCAN -> Density Based Spacial Clustering of Application with Noise

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
```

```python
X, y = make_moons(n_samples=500, shuffle=True, noise=.1, random_state=None)
```

```python
plt.scatter(X[:,0], X[:,1],  c =y )
```

```python
sns.scatterplot(x = X[:,0], y = X[:,1], hue= y)
```

### for this data set if i apply kmeans it will fail to perform clusters 

### Thus i will be using DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
# Core Point -> if you take a circle, if it has 4+ more point is core point 
# Border Point -> if you take a circle if it has less than 4 point its called border point 
# noise -> if there are no point in that radious its called Noie 
```

```python
dbs = DBSCAN(eps=0.15, min_samples= 3 )
```

```python
dbs.fit(X)
```

```python
y_pred = dbs.fit_predict(X)
```

```python
sns.scatterplot(x = X[:,0], y = X[:,1], hue= y_pred , )
```

```python
plt.scatter(x = X[:,0], y = X[:,1], c= y_pred)
```
