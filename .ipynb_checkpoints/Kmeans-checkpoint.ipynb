```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples= 500, n_features=2, centers= 5,random_state=3)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

### The goal is to cluster the blobs 

```python
#i,e :
plt.scatter(X[:,0], X[:,1], c = y)
plt.show()
```

## Algorithem 

# Step 1 < Randomly initialize the the centers >

```python
# k is number of centers 

k = 5 

# randomly give clor 

color = ['green','yellow','blue','orange','red']

# Dictionary of clusters :- 
    # this clusters dictionary will contain the data point and the color and center location 
clusters = {}

for i in range(k):
    # initialize center location for c<i> cluster
    center = np.random.uniform(low = -8, high = 8, size = 2)
    points = []
    
    cluster = {
        "center": center,
        "points": points,
        "color" : color[i]
    }
    clusters[i] = cluster
```

```python
clusters
```

## Step 2, Assign Data point to the nearest cluster 

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assignPointsToCluster():
    
    # iterate over data point 
    for i in range(X.shape[0]):
        dist = []
        for j in range(k):
            d = distance(clusters[j]['center'], X[i])
            dist.append(d)
        
        nearest_cluster = np.argmin(dist) # the argmin gives the positino thus acting as cluster id
        
        clusters[nearest_cluster]['points'].append(X[i])
            
            
```

```python
assignPointsToCluster()
```

```python
clusters
```

```python
def plotClusters():
    
    for cx in range(k):
        pts = np.array(clusters[cx]['points'])
        cluster_color = clusters[cx]['color']
        
        center = np.array(clusters[cx]['center'])
        
        try:
            plt.scatter(pts[:,0],pts[:,1] , c = cluster_color)
            plt.scatter(center[0],center[1], c = 'black',  s = 300, marker= "*" )
            
        except:
            pass
        
        
        
        
    plt.show()
    
```

```python
plotClusters()
```

# Step 3: Update Clusters 

```python
def updateClusters():
    for cx in range(k):
        pts = clusters[cx]['points']
        
        pts = np.array(pts)
        mean = np.mean(pts, axis = 0)
        
        # update the cente value
        clusters[cx]['center'] = mean
        
        # empty the list 
        clusters[cx]['points'] = []
```

```python
updateClusters()
```

```python
clusters
```

```python
## plot the new 
```

```python
plotClusters()
```

```python
## now we need to repeat the step 2 and 3 
```

```python

assignPointsToCluster()
plotClusters()
updateClusters()
```

```python
assignPointsToCluster()
plotClusters()
updateClusters()
```
