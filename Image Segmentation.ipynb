```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
image = plt.imread("./image.jpg")
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
#reshape
r_image = image.reshape(-1, 3)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(r_image)
```

```python
#Cluster Centers
centers = model.cluster_centers_
centers
```

```python
## to plot all colors 
```

```python
arr = np.zeros((100,100,3), dtype = "int")
```

```python
arr[:,:,:] = centers[0]
plt.imshow(arr)
```

```python
arr[:,:,:] = centers[1]
plt.imshow(arr)
```

```python
arr[:,:,:] = centers[0]
plt.imshow(arr)
```

## image Segmentation 

```python
model.labels_
```

```python
new_image = np.zeros((1384*2400, 3) )
```

```python
new_image.shape
```

```python
## now i will replace the values of the zeos in new image with the center colors 
```

```python
for i in range(new_image.shape[0]):
    new_image[i] = centers[model.labels_[i]]
```

```python
new_image = new_image.reshape(image.shape)
```

```python
plt.imshow(new_image/255)
plt.title("compressed image")
plt.show()
```
